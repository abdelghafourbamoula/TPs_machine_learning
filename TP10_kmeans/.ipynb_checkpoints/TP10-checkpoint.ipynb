#### ***Q1***-utilisez la fonction ***make_blobs(n_samples,centers, cluster_std,random_state)*** du module ***sklearn.datasets*** pour generer trois classes puis visualiser ces classes 



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=200, centers=3, cluster_std=0.8, random_state=0)
X.shape, y.shape
```

```python
sns.scatterplot(X[:,0], X[:,1], c=y)
```

#### ***Q2***  Ecrire un script pour pr√©dire la classe de x avec l'algorithme du k-means puis visualiser ces trois classe avec les cluster_centers

```python
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cl = KMeans(n_clusters=3)
cl.fit(X)
y_pred = cl.predict(X)
```

```python
accuracy_score(y, y_pred)
```

```python
cm = confusion_matrix(y, y_pred)
sns.heatmap(cm, annot=True)
```

```python
centroids = cl.cluster_centers_
sns.scatterplot(X[:,0], X[:,1], c=y)
plt.scatter(centroids[:,0], centroids[:,1], color='red', lw=3)
```

#### Q3. A quelle classe apartient les trois points suivants: x1=(-2,1); x2(0,1) ;x3=(2.5,5)  

```python
test = np.array([[-2,1],
                 [0,1],
                 [2.5,5]])
predictions = cl.predict(test)

for i in range(len(test)):
    print('- X',i,' appartient au class ',predictions[i])
```

```python
sns.scatterplot(X[:,0], X[:,1], hue=y)
plt.scatter(test[:,0], test[:,1], color='blue', lw=3)
```

```python
 from sklearn.datasets import load_iris
```

```python
data = load_iris()
X = data.data[:,:2]
Y = data.target
sns.scatterplot(X[:,0],X[:,1],c=Y)
```

```python
cl = KMeans(n_clusters=3)
cl.fit(X)
```

```python
hist = []
for i in range(1,30):
    cl = KMeans(n_clusters=i)
    cl.fit(X)
    hist.append(cl.inertia_)
# print(hist)
plt.plot(hist)
```
